```python
def flat_list(a):
    flatlist = []
    for element in a:
        if type(element) is list :
            for item in element:
                flatlist.append(item)
        elif type(element) is dict:
            l1 = list(element.items())
            for i in l1:
                if type(i) is tuple:
                    for tup in i:
                        if type(tup) is list:
                            for j in tup:
                                flatlist.append(j)
                        elif type(tup) is tuple:
                            for k in tup:
                                flatlist.append(k)
        elif type(element) is tuple:
            for i in element:
                flatlist.append(i)
        elif type(element) is set:
            for i in element:
                flatlist.append(i)
        else:
            flatlist.append(element)
    print(flatlist)
    product  = 1
    for i in flatlist:
        if type(i) is int:
            product = product*i
        else:
            continue
    print(product)
```

```python
flat_list([1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,
22, 61, 34)}, [56, 'data science'], 'Machine Learning'])
```

```python
sen = input("enter the sentence")
str1 = ""
sen1 = sen.lower()
for i in sen1:
    if i == "a":
        str1 += "z"
    elif i == "b":
        str1 += "y"
    elif i == "c":
        str1 += "x"
    elif i == " ":
        str1 += "$"
    else:
        str1 += i
print(str1)
```
