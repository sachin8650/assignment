```python
marks = int(input("enter the marks that you obtained"))
if marks>0 and marks <=100:
    if marks>90:
        print("GRADE - A")
    elif marks>80 and marks <=90:
        print("GRADE - B")
    elif marks>60 and marks<=80:
        print("GRADE - C")
    else:
        print("GRADE - D")
else:
    print("enter correct marks")
```

```python
cost = int(input("enter the cost of the bike"))
if cost > 0:
    if cost > 100000:
        print("15%")
    elif cost>50000 and cost <=100000:
        print("10%")
    else:
        print("5%")
```

```python
city = input("enter the name of any city from the following :\n Delhi \t Agra \t Jaipur")
if city == "Delhi":
    print("Red Fort")
elif city == "Agra":
    print("Taj Mahal")
elif city == "Jaipur":
    print("Jal Mahal")
else:
    print("Please enter the correct name of city")
```
