```python
password = input("enter the password")
up =0
low = 0
num = 0
char = 0
for i in password:
    if i.islower():
        low = low +1
    elif i.isupper():
        up +=1
    elif i.isnumeric():
        num = num +1
    else:
        char += 1
if up >=2 and low >=2 and num >=1 and char >=3 and len(password)>=10:
    print("Valid Password")
else:
    print("Invalid Password")
        
```

```python
str1 = input("enter the string")
str2 = list(str1)
counter = 1
for i in str2:
    if type(i) is int:
        continue
    else:
        counter = 0
if counter == 1:
    print("string is numeric")
else:
    print("string is not numeric")
```
